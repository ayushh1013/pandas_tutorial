```python
 np.linspace(10,12,11)
```

```python
def give_linspace(first,last,nums):
    all_vals = []
    a1 = first
    an = last 
    n = nums
    d = (an - a1)/(n-1)
    for i in range(1,nums + 1):
        next_num = a1 +(i-1)*d
        all_vals.append(next_num)
    return all_vals
```

```python
print(give_linspace(1,5,10))
```

```python

python_app = np.array(give_linspace(1,50,15))
```

```python
nump = np.linspace(1,50,15)
```

```python
python_app == nump
```
